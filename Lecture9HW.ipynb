```python
# 1. How do you import the pandas package in Python?
import pandas as pd
```

```python
# 2. What function do you use to read a CSV file in pandas?

df = pd.read_csv('https://raw.githubusercontent.com/juanklopper/TutorialData/main/heart_failure.csv')
```

```python
# 3. How do you display the first 5 rows of df in pandas?
print(df.head())

```

```python
# 4. How do you calculate the mean of a column named Age in a DataFrame named df?
print(df['age'].mean())
```

```python
# 5. How do you calculate the median of a column named creatinine_phosphokinase in a DataFrame named df?
print(df['creatinine_phosphokinase'].median())
```

```python
# 6. How do you calculate the standard deviation of a column named ejection_fraction in a DataFrame named df?
print(df['ejection_fraction'].std())
```

```python
# 7. How do you find the number of missing values in each column of a DataFrame named df?
print(df.isnull().sum())

```

```python
# 8. How do you calculate the correlation between two columns, age and serum_creatinine, in a DataFrame named df?
print(df['age'].corr(df['serum_creatinine']))
```

```python
# 9. How do you select a subset of a DataFrame df where the column Age is greater than 60?
print(df[df['age'] > 60])
```

```python
# 10. How do you calculate the range (maximum - minimum) of a column named serum_sodium in a DataFrame named df?
print(df['serum_sodium'].max() - df['serum_sodium'].min())

```

```python
# 11. How do you group a DataFrame df by a column named sex and calculate the mean of serum_sodium within each group?
df.groupby('sex').serum_sodium.mean()

```

```python
# 12. How do you group a DataFrame df by two columns, Department and Job Title, and count the number of rows within each group?
df.groupby(['age', 'serum_sodium']).size()
```

```python
# 13. How do you use the groupby method to find the maximum Age in each serum_sodium in a DataFrame df?
df.groupby('serum_sodium').age.max()
```

```python
# 14. How do you create a cross-tabulation table that shows the frequency count of Age (rows) and serum_sodium (columns) in a DataFrame df?
pd.crosstab(df['age'], df['serum_sodium'])
```

```python
# 15. How do you create a cross-tabulation table that shows the mean age for each combination of serum_sodium (rows) and serum_creatinine (columns) in a DataFrame df?
pd.crosstab(df['serum_sodium'], df['serum_creatinine'], values=df['age'], aggfunc='mean')
```
